```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
df
```

```python
import pandas as pd
import numpy as np

# Check data types to identify non-numeric columns
print(df.dtypes)

# Convert categorical data (if any) to numeric (like 'Yes'/'No' â†’ 1/0)
df.replace({'Yes': 1, 'No': 0}, inplace=True)

# Select only numeric columns to avoid conversion errors
df_numeric = df.select_dtypes(include=[np.number])

# Display the filtered numeric DataFrame
display(df_numeric)  # This works well in Jupyter Notebook

# Calculate and display covariance matrix
cov_matrix = df_numeric.cov()
display(cov_matrix)
```

```python
# Compute and display the correlation matrix
corr_matrix = df_numeric.corr()
display(corr_matrix)
```
